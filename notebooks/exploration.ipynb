```python
df=df.dropna()
df.isnull().sum().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

top10_unique=df.sort_values("popularity",ascending=False).drop_duplicates(subset="track_name").head(10)

plt.figure(figsize=(10,6))
sns.barplot(data=top10_unique,x="popularity",y="track_name",hue="track_name",palette="coolwarm",legend=False)
plt.title("Top 10 Most Popular Unique Tracks")
plt.xlabel("Popularity score")
plt.ylabel("Track Name")
plt.tight_layout()
plt.show()
```

```python
top_genres=df["track_genre"].value_counts().head(10).index
print(top_genres)
```

```python
df_genres=df[df["track_genre"].isin(top_genres)]
print(df_genres.shape)
```

```python
df_genres["track_genre"].value_counts()
```

```python
genre_features=df_genres.groupby("track_genre")[["energy","danceability","acousticness"]].mean()
print(genre_features)
```

```python
genre_features.plot(kind='bar',figsize=(12,6),colormap='viridis')
plt.title("Average Music features by Genre")
plt.xlabel("Genre")
plt.ylabel("Average Score(0-1)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
